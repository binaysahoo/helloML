```python
np.max(df11)
```

```python
def band(rec):   
    """
    band = max - min of the column
    """
    return rec.max() - rec.min()


#band(df11)
df11.apply(band)   # columns wise max - min
```

```python
# row wise : band : max - min
df11.apply(band,axis=1) 
```
