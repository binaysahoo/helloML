```python
#===================================================
# selection 
#===================================================
# 1.  filter of column or columns
# select age from table 
print(df['age'])    # or df.age

print("Type of data:",type(df['age']))
_ages = df.age.values
ages = list(_ages)
print("ages:",_ages)
print("type of _ages:",type(_ages))
print("type of ages:",type(ages))
```

```python
# 1.1  filter of columns
# select name,age from table
df[['name','age']]
#df.loc[:,['name','age']]
```

```python
# 2. filter of rows
#   select * from table where vehicle is NA
df[df.vehicle == 'NA']
```

```python
# 2.1 select all rows where age < 40 and vehicle not NA
df[(df.age <= 40 ) & (df.vehicle != 'NA')]
```

```python
# filter rows and columns 
# 3 . select name,age,city,vehicle where where age < 40 and vehicle not NA
#df[(df.age <= 40 ) & (df.vehicle != 'NA')].loc[:,['name','age','city']]
df[['name','age','city','vehicle']][(df.age <= 40 ) & (df.vehicle != 'NA')]
```

```python
# group by vehicle
# select count(*) from table group by vehicle 
df.groupby('vehicle').count()   # max(),sum(),mean(),median
```

```python
# # select count(*) from table group by vehicle and city  
df.groupby(['vehicle','city']).count()
```
